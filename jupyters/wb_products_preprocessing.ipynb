```python
import numpy as np

x = np.random.rand(50)
x
```

```python
def formula(x):
    delta_T = 1/64

    sum_1 = 0
    for i in range(1, len(x)-1):
        value_1 = pow(
            ((x[i+1] - 2*x[i] + x[i-1])
             / pow(delta_T, 2)),
            2
        )

        sum_1 += value_1

    sum_2 = 0
    for i in range(1, len(x)-1):
        value_2 = pow(
            ((x[i-1] - x[i])
             / delta_T),
            2
        )

        sum_2 += value_2

    return sum_1 + sum_2/2
```
